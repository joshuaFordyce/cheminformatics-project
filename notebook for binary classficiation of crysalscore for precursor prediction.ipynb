```python
from sklearn.model_selection import train_test_split
```

```python
X_train_full, X_valid_full, y_train, y_valid = train_test_split(X, y, train_size = .4, test_size = .6, random_state = 0)

X_training = X_train_full

X_validy = X_valid_full
```

```python
perksovites_model = DecisionTreeClassifier(criterion = 'entropy',random_state = 0)

perksovites_model.fit(X_training, y_train)
```
